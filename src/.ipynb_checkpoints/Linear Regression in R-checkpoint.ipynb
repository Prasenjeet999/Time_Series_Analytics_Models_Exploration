## 1. Regression Background
******************************************************************

### 1. Intution

Let's assume, there are some N random Variables:

$ Y_1, Y_2, ..., Y_n -idd- N(\mu, \sigma^2)$


Now, we are considering cases for fixed $\mu$ but what if the mean is not fixed but a function!


Let's consider:


$Y = f(X\beta)$       

Given, Y is a linear function of $\beta$s, in such cases, we call it Linear Regression.

**If some assumptions are met and Y is in a linear relationship with some Constants ( $X, \beta$ ), the function is said to be Linear Regression.**


- $Simple \space Linear \space Regression: Y = \beta_0 + \beta_1.X + \epsilon$


- $Multiple \space Linear \space Regression: Y = \beta_0 + \beta_1.X_1 + \beta_2.X_2 + ... + \beta_p.X_p + \epsilon$


Also, often, while taking measurements for Xs, there will always be some measurement error which is denoted by $\epsilon$

---------------------------------------------------------------------------------------


### 2. Matrix Representation :

`Hypothetical Equation` for best fit line  (what we wish to achieve): 

$\mathbf{Y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}$,

-OR-

$\begin{equation} Y_i = \beta_0 + \sum^p_{j=1}\beta_j x_{i,j} + \varepsilon_i,\end{equation}$ for $i = 1,...,n$, where $n$ is the number of data points (measurements in the sample), and $j = 1,...,p$, where

- $p+1$ is the number of parameters in the model.


- $Y_i$ is the $i^{th}$ measurement of the response variable.


- $x_{i,j}$ is the $i^{th}$ measurement of the $j^{th}$ predictor variable.


- $\varepsilon_i$ is the $i^{th}$ error term and is a random variable, often assumed to be $N(0, \sigma^2)$.


- $\beta_j$, $j = 0,...,p$ are unknown parameters of the model. We hope to estimate these, which would help us characterize the relationship between the predictors and response.

Matrix Formulation:

$\begin{bmatrix} y_{1} \\ y_{2} \\ \vdots \\ y_{n} \end{bmatrix} = \begin{bmatrix} 1 & x_{1,1} & x_{1,2} & ... & x_{1,p} \\ 1 & x_{2,1} & x_{2,2} & ... & x_{2,p} \\ \vdots & \ddots & \ddots & \vdots \\ 1& x_{n,1} & x_{n,2} & ... & x_{n,p} \end{bmatrix} * \begin{bmatrix} \beta_{0} \\ \beta_{1} \\ \vdots \\ \beta_{p} \end{bmatrix} + \begin{bmatrix} \epsilon_{1} \\ \epsilon_{2} \\ \vdots \\ \epsilon_{n} \end{bmatrix}$



where:

- Response Matrix (Y) : $\boldsymbol{Y} = {[} \begin{matrix} y_1 & y_2 & y_3 & ... & y_n\end{matrix} {]}^T$
- $dim(Y) = n * 1$

- Predictor Matrix (X): $\boldsymbol{X} = \begin{bmatrix} 1 & x_{1,1} & x_{1,2} & x_{1,3} & ... & x_{1,p} \\ 1 & x_{2,1} & x_{2,2} & x_{2,3} & ... & x_{2,p} \\ & ... & \\1 & x_{n,1} & x_{n,2} & x_{n,3} & ... & x_{n,p} \end{bmatrix}$
- $dim(X) = n * (p+1)$
- Parametric Matrix (beta): $\boldsymbol{\beta} = {[} \begin{matrix} \beta_0 & \beta_1 & ... & \beta_p\end{matrix} {]}^T$
- $dim(Y) = (p+1) * 1$
- Error Term Matrix (e): $\boldsymbol{\varepsilon} = {[} \begin{matrix} \varepsilon_1 & \varepsilon_2 & ... & \varepsilon_n\end{matrix} {]}^T$
- $dim(\varepsilon) = n*1$

---------------------------------------------------------------------------------------


### 3. Estimating $\beta$  (i.e. $\hat \beta$):

Four common methods:- 

- 1. Ordinary Least Squares (OLS)
- 2. Maximum Likelihood Estimation (MLE)
- 3. Generalized Least Squares (GLS)
- 4. Gradient Descent Optimzation (GDO)


Using OLS, or MLR  we basically wish to estimate $\hat \beta$ such that the *SSE (Sum of Squared Errors/Residuals)* is **minimized**, and we get the closest line possible. $\widehat \beta$  could be shown as:  



- $\widehat Y = X\widehat \beta$  


- Where, $\widehat \beta = (X^TX)^{-1}X^TY$  

---------------------------------------------------------------------------------------


### 4. Estimating $H$ Matrix:

$H$ matrix is just another name for the projection matrix and find it's trace:

$$\widehat Y = HY$$

$=> \widehat Y = X \widehat \beta$     
$=> \widehat Y = X (X^TX)^{-1}X^TY$  
$=> \widehat Y = HY$  

$=> H = X(X^TX)^{-1}X^T$ 


Now this is your final H matrix.

Here, let $A = X(X^TX)^{-1}$ and $B = X^T$ 

Applying commutative property of the trace operator: $tr(AB) = tr(BA)$ which could be easily derived (*just try multiplying two matrices and summing their diagonals*): 

$trace(H) = tr(X (X^TX)^{-1} X^T) = tr(AB) = tr(BA) = tr(X^T  . X (X^TX)^{-1} ) = tr(I_{X^TX}) = tr(I_{(p+1) \times n . n \times (p+1)} = tr(I_{p+1 \times p+1}) = tr(I_{p+1}) = tr(I_{p+1}) = \sum^{p+1}_{i=1} (1) = p+1$  

Therefore, **$trace(H) = p+1$**

---------------------------------------------------------------------------------------


### 5. Distrbituons and Sampling Distributions:

#### 5.1. ERROR TERM ~

$$\boldsymbol{\varepsilon_i} \sim N(\mathbf{0},\sigma^2I_n)$$

It can be assumed that the model is following all the regular assumptions and it tells us that each element of the Error Vector would be independently and identically distributed (iid) with mean equal to $0$ and variance $\sigma^2$.  This means, that for each error vector element, i.e. $\varepsilon_i \sim N(0,\sigma^2)$   such that,  $i=1,2,...,n$

$E(\varepsilon) = E \left(\begin{matrix} \varepsilon_0 \\ \varepsilon_1 \\ ... \\\varepsilon_n \end{matrix}\right) = \left(\begin{matrix} E(\varepsilon_0) \\ E(\varepsilon_1) \\ ... \\E(\varepsilon_n) \end{matrix}\right) = \left(\begin{matrix} 0 \\ 0 \\ ... \\ 0 \end{matrix}\right) = 0 $

$Var(\varepsilon) = Var\left(\begin{matrix} \varepsilon_{0,0} & \varepsilon_{0,1} & \cdots & \varepsilon_{0,n} \\ \varepsilon_{1,0} & \varepsilon_{1,1} & \cdots & \varepsilon_{1,n} \\ \vdots & \vdots & \ddots & \vdots \\ \varepsilon_{n,0} & \varepsilon_{n,1} & \cdots & \varepsilon_{n,n} \\ \end{matrix}\right) = \left(\begin{matrix} \sigma^2 & 0 & \cdots & 0 \\ 0 & \sigma^2 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & \sigma^2 \\ \end{matrix}\right) = \sigma^2 \left(\begin{matrix} 1 & 0 & \cdots & 0 \\ 0 & 1 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & 1 \\ \end{matrix}\right) = \sigma^2I_n$

Since the expectation and variance of each element of the error matrix are $\mathbf{0}$ and $\sigma^2I_n$, respectively, we can conclude that $\varepsilon_i \sim N(0,\sigma^2)$ for any $i=1,2,...,n$ 

.

#### 5.2. RESPONSE (Y) ~

$$\mathbf{Y} \sim N(X\boldsymbol\beta, \sigma^2I_n)$$:

Since, it's understood that error term follows normality, with $\varepsilon_i \sim N(0,\sigma^2)$ and $X, \beta$ are constants:

$E[Y] = E[X\beta + \varepsilon] = E[X\beta] + E[\varepsilon] = X\beta.E[1] + E[\varepsilon] = X\beta + 0 = X\beta$

Since, we know:-  $Var[A] = E[A^2] - E[A]^2$

$Var[Y] = Var[X\beta + \varepsilon] = E[(X\beta + \varepsilon)^2] - E[X\beta + \varepsilon]^2$

$= E(X^2\beta^2 + 2X\beta\varepsilon + \varepsilon^2) - (E[X\beta] + E[\varepsilon])^2$

$=X^2\beta^2E[1] + 2X\beta E[\varepsilon] + E[\varepsilon^2] - X^2\beta^2E[1]^2 - E[\varepsilon]^2$

$=X^2\beta^2 - X^2\beta^2+ 0 + E[\varepsilon^2]  - E[\varepsilon]^2$

$=E[\varepsilon^2]  - E[\varepsilon]^2$

$=Var[\varepsilon] = \sigma^2I_n$

Therefore, $\mathbf{Y} \sim N(X\boldsymbol\beta, \sigma^2I_n)$

.

#### 5.3. Estimates of Parameters ($\widehat\beta$) ~

For the linear regression model $\mathbf{Y} = X\boldsymbol\beta + \boldsymbol{\varepsilon}$ 

- Regression parameters does not have sampling distributions.
- Estimators of regression parameters have sampling distributions.


$$\hat \beta \sim N(\boldsymbol\beta,  \sigma^2(X^TX)^{-1} )$$

$\boldsymbol{\widehat\beta}$ is the best linear unbiased least squares estimator of $\boldsymbol{\beta}$, which means on average $\boldsymbol{\widehat\beta}$ gets a perfect estimate for parameters.

** It $\widehat\beta$ has the lowest variance among unbiased estimators. For biased estimators, they could have lower variance. **



Since, $\hat \beta = (X ^TX)^{-1} X^T Y$, and under Gauss Markov Assumptions, $E[Y] = X\beta$

$E[\hat\beta] = E[(X^TX)^{-1} X^T Y] = (X^TX)^{-1} X^T * E[Y] = (X^TX)^{-1} X^T * X\beta = (X^TX)^{-1} X^TX * \beta = 1 * \beta = \beta $



Since for a given linear equation, with A being a constant in $Z = AY$: 

$Var[Z] = A.Var[Y].A^T$ 

Then,

$Var[\hat\beta] = Var[(X^TX)^{-1} X^T Y]$

$= (X^TX)^{-1} X^T . Var[Y] . [(X^TX)^{-1} X^T]^T$

$= (X^TX)^{-1} X^T . \sigma^2 . [X . (X^TX)^{-1}]$

$=\sigma^2 . (X^TX)^{-1} X^T X (X^TX)^{-1}$

$= \sigma^2 . (X^TX)^{-1}$



#### 5.4. Estimates of Response  ($\widehat Y$) ~


$$\hat Y \sim N(X\boldsymbol\beta,  \sigma^2.H )$$

$=> E[\widehat Y] = E[X \widehat \beta] = X E[\widehat \beta] = X\beta$     


$=> Var[\widehat Y] = Var[X \widehat \beta] = X.Var[\widehat \beta].X^T =   X. \sigma^2(X^TX)^{-1}. X^T = \sigma^2 . X(X^TX)^{-1}X^T = \sigma^2.H$  

.

#### 5.5. Estimate of Unknown Error Variance:


Since, $\sigma^2$ is unknown, we can estimate it. The unbiased estimator is  $\hat \sigma^2$

$\widehat \sigma^2 = \frac{RSS}{n-p-1}$     is a unbiased estimator.

.

#### 5.6. Estimate of Error Term (residual):

Since, $$Error \space \epsilon \sim N(0, Constant Variance) \sim N(0, \sigma^2)$$

We can estimate the error term (residuals):

$$ Residuals \space  \widehat \epsilon \sim N( 0, \sigma^2(I_n - H) \space)$$ - which has a non-constant variance.

---------------------------------------------------------------------------------------


### 6. Covariance of Parameters:

Each sample will give one $\hat\beta$ represented by $\hat \beta_i$. 

It can be shown that, $\hat \beta_i$, $\hat \beta_j$ are correlated!

### 7. Explanation vs Prediction:

- Explanation: Any analysis done on training data that involves determining casual mechanisms is Explanation.
- Prediction: Any analysis done on testing data that does not involve any casual inference.

NOW,

#### 7.1. Explanation (on `train` data):

$ \hat Y = X . \hat \beta$      ; where X - is the train data

- Use: fitted(lm) or predict(lm)

Matrix Formulation:

$\begin{bmatrix} \hat y_{1} \\ \hat y_{2} \\ \vdots \\ \hat y_{n} \end{bmatrix} = \begin{bmatrix} 1 & x_{1,1} & x_{1,2} & ... & x_{1,p} \\ 1 & x_{2,1} & x_{2,2} & ... & x_{2,p} \\ \vdots & \ddots & \ddots & \vdots \\ 1& x_{n,1} & x_{n,2} & ... & x_{n,p} \end{bmatrix} * \begin{bmatrix} \hat \beta_{0} \\ \hat \beta_{1} \\ \vdots \\ \hat \beta_{p} \end{bmatrix}$


#### 7.2. Prediciton:  (on `test` data):   

$ \hat Y^{\star} = X^{\star} . \hat \beta + \epsilon$      ; where X_star - is the test data

- Use: predict(lm, newdata = test)

### Merging three datasets into file

```r
convert_to_knowns <- function(df, date_format) {
    
    # Factorizating
    df$Store <- factor(df$Store)
    df$Dept <- factor(df$Dept)
    df$Type <- factor(df$Type)
    df$Size <- as.numeric(df$Size)

    # Converting timestamp to date format
    df$Date = as.Date(df$Date, format = date_format)
    df = arrange(df, Date, decreasing=TRUE)
    return(df)
}
```

# 6. Linear Regression Modelling

```r
# Plot Functions for Simple Linear Regression Analysis
# Y = beta_0 = beta_1.X

#############################################################################
#      Simple Linear Regression with X,Y
#
#  > plot_slr(X=marketing$facebook, Y=marketing$sales)
#############################################################################
plot_slr <- function(X,Y) {
    data = data.frame(x=X, y=Y)
    ggplot(data, aes(x=x, y=y))+
        geom_point() +
        geom_smooth(method="lm", col="red") +
        stat_regline_equation(label.x = 0, label.y = 0) +
        labs(title = "Simple Linear Regression", x = "X Predictor", y = "Y Response")
}

#############################################################################
#      Simple Linear Regression with fitted lm()
#
#  > plot_slr_model(slr_lm_model)
#############################################################################
plot_slr_model <- function(lm_model) {
    if(length(lm_model$coeff) == 2){
        ggplot(lm_model$model, aes_string(x = names(lm_model$model)[2], y = names(lm_model$model)[1])) + 
          geom_point() +
          stat_smooth(method = "lm", col = "blue") +
          labs(title = paste(
              "\n R2 = ", round(summary(lm_model)$r.squared, 5),
              "\n Adj R2 = ", round(summary(lm_model)$adj.r.squared, 5),
              "\n Intercept =", round(lm_model$coef[[1]],5 ),
              "\n Slope =", round(lm_model$coef[[2]], 5),
              "\n P =", round(summary(lm_model)$coef[2,4], 5)))
    } 
    else{
        print("Invalid LR model - pass a Simple LR model only.")
    }
}

#############################################################################
#      Simple Linear Regression with given beta0, beta1
#
#  > plot_slr_coeff(X=marketing$facebook, Y=marketing$sales, beta_0, beta_1)
#############################################################################
plot_slr_coeff <- function(X, Y, beta_0, beta_1) {
    data = data.frame(x=X, y=Y)
    ggplot(data = data, aes(x = x, y = y)) +
        geom_point() +
        geom_abline(intercept = beta_0, slope = beta_1, col="blue")
}
```

## 6.1. LR By Hand

```r
head(train)
```

$n, p$

```r
n = nrow(train)
p = ncol(train) - 1

cat("n =", n, "<--> p =", p)
```

X, Y

\begin{bmatrix} 1 & x_{1,1} & x_{1,2} & ... & x_{1,p} \\ 1 & x_{2,1} & x_{2,2} & ... & x_{2,p} \\ \vdots & \ddots & \ddots & \vdots \\ 1& x_{n,1} & x_{n,2} & ... & x_{n,p} \end{bmatrix} 

```r
# Design Matrix ::
X = cbind(1, train$youtube, train$facebook, train$newspaper)

# Target/Response
y = train$sales
```

$\beta$

```r
# Estimating Beta parameter using OLS method ::

# Solution: B = (XT.X)^-1 . XT.Y

# Gram Matrix: XT.X
# Inverse should exist, OLS has a unique solution, If it doesn't exist that means there is some lineary dependency among predictors such that, one predictor could be represented as a linear combination of other predictors. 
XTX = t(X)%*%X

# Checking if inverse of Gram matrix exists or not using determinant of XTX:
if (abs(det(XTX)) > 1e-10) {
  cat("Gram Matrix Inverse exists!\n")
  
  # Calculating Inverse
  XTX_inverse = solve(XTX)
  
  # Calculating Beta:
  beta = XTX_inverse %*% t(X) %*% y

} else {
  cat("Gram Matrix Inverse DOES NOT exist! OLS cannot be used here!\n")
}

# By hand solution of Beta
beta.intercept = beta[1]
beta.youtube = beta[2]
beta.facebook = beta[3]
beta.newspaper = beta[4]
cat("Intercept Parameter:", beta.intercept,
    "\nParameter for Youtube:", beta.youtube,
     "\nParameter for Facebook:", beta.facebook,
     "\nParameter for Newspaper:", beta.newspaper)

# OR
# beta = solve(t(X)%*%X) %*% t(X) %*% y
```

$\widehat Y$

```r
# Prediction ::

# Y_hat = X . Beta_Hat
y_hat = X %*% beta
```

$Projection \space Matrix \space (H)$

Properties of H (Orthogonal Projection):

- H is a symmetric matrix.


- $HH = H$


- $H^k = H$, where $H^k$ is $H$ multiplied by itself $k$ times.


- $tr(H) = p+1$ where $p$ is the number of predictors in the regression model and $tr(\cdot)$ is the "trace" of the the matrix, i.e., the sum of the diagonals.


- $H\mathbf{Y} = \widehat{\mathbf{Y}}$


- $H\widehat{\mathbf{Y}} = \widehat{\mathbf{Y}}$

```r
# By hand

H_matrix = X %*% solve(t(X) %*% X) %*% t(X)

# tr(H) = p+1 = 4
trace_H = sum(diag(H_matrix))

cat("Computed H-Matrix. Trace(H) by hand: ", trace_H)
```

$Metrics$

```r
## Metrics ::

# Residual: (y - y_hat)
r = y - y_hat

# ResidualSS
rss = sum((y - y_hat)^2)
cat("\n:: METRICS ::\n\nRSS: ", rss)

# ExplainedSS
ess = sum((y_hat - mean(y))^2)
cat("\nESS: ", ess)

# TotalSS
tss = sum((y - mean(y))^2)
cat("\nTSS: ", tss)
cat("\n\nTSS = RSS + ESS = ", rss + ess)

# Mean Squared Error = RSS/n
mse = 1/n * sum((y - y_hat)^2)
cat("\n\nMSE = ", mse)

# Root MSE
rmse = sqrt(mse)
cat("\nRMSE = ", rmse)

# Mean Abs Error
mae = 1/n * sum(abs(y - y_hat))
cat("\nMAE = ", mae)

# R Squared
r2 = 1 - (sum((y - y_hat)^2) / sum((y - mean(y))^2))
cat("\n\nR Squared = ", r2)

# Adjusted R Squared
# - beacuse if we increase p (num of predictors), R2 will keep on increasing, so we add penalty!
# n = dim(adv)[1], p = 3
adj.r2 = 1 - ( (1-r2) * (n-1) / (n-3-1) )
cat("\nAdjusted R Squared = ", adj.r2)
```

### Sampling Dist of $\widehat \beta$


The estimators of regression parameters have a sampling distribtuion, bivariate normal dist.

$\hat \beta \sim N(\mu, \sigma^2)$ 

- $\mu = Expectation \space [\widehat \beta_j] = \beta_j$


- $\sigma^2 = Variance \space [\widehat \beta_j] = \widehat \sigma^2 (X^T X)^{-1}$


- $\sigma = Standard\space Deviation \space [\widehat \beta_j] = Standard \space Error = \sqrt{ \widehat \sigma^2 [(X^T X)^{-1}]_{jj}}$


where,

- Estimate of error variance is given by: $\widehat \sigma^2 = \frac{RSS} {(n-(p+1))}$
- $jj$ means diagonal values of standard error matrix.

```r
# assumed default value
exp_beta = 0

cat("Degress of Freedom = ", n-p-1)
```

```r
# Est_variance = RSS / n - (p+1)
estimated_variance = rss/(n - (p+1))
estimated_variance

# Standard Error (aka overall standard deviation)
sde_matrix = sqrt(estimated_variance * solve(t(X)%*%X))
sde_matrix
```

Standard Error = Average of Std from n samples.

- Matching with lm() values.
- Diagonals (jj) should tell us the variance for each parameter estimate.

t-Statistics:

$t(stat) = \frac{\hat \beta_j} {S.E(\hat \beta_j} \sim t.dist_{n-p-1}$

```r
# t-value for intercept
t_stat.intercept = beta.intercept/sde_matrix[1,1]      # jj - diagonals from standard error matrix

# t-value for intercept
t_stat.youtube = beta.youtube /sde_matrix[2,2]

# t-value for intercept
t_stat.facebook = beta.facebook/sde_matrix[3,3]

# t-value for intercept
t_stat.newspaper = beta.newspaper/sde_matrix[4,4]

cat("T-Stats:\n\nt-stat Intercept:", t_stat.intercept,
    "\nt-stat for Youtube:", t_stat.youtube,
     "\nt-stat for Facebook:", t_stat.facebook,
     "\nt-stat for Newspaper:", t_stat.newspaper)
```

F-Statistics:

$F(stat) = \frac{\frac {RSS_{reduced} - RSS_{full}}{n-q-1 - (n-p-1) = p-q}}{\frac{RSS_{full}}{n-p-1}} \sim F.dist_{p-q,n-p-1}$

### 6.2.5. Interval Estimates for `Prediction`

- A). Confidence and Prediction intervals on training data
- B). Confidence and Prediction intervals on testing data

#### A. Over Training Data ($\hat Y_j$)

Given, for confidence:

$$ \hat Y  = X \hat \beta$$

- $E[\hat Y] = E[X\hat \beta] = X\beta$


- $Var[\hat Y] = Var[X \hat\beta] = X.Var[\hat \beta].X^T = X.\hat \sigma^2(X^TX)^{-1}.X^T = H \hat \sigma^2$


- $ \hat Y \sim N(X\beta, \space \space \sigma^2.H)$


Confidence Interval Estimate:-
$$ \hat Y_j \space +- \space \space CriticalValue \space . \space SE(\hat Y_j)$$
$$ \hat Y_j \space +- \space \space t.stat_{n-p-1} . \sqrt{\sigma^2H}$$

$$ Lower.bound: \hat Y_j - t.stat_{n-p-1} . \sqrt{\sigma^2.X_{train}.(X_{train}^T.X_{train})^{-1}.X_{train}^T}$$
$$ Upper.bound: \hat Y_j + t.stat_{n-p-1} . \sqrt{\sigma^2.X_{train}.(X_{train}^T.X_{train})^{-1}.X_{train}^T}$$

.

Given, for prediction:

$$ \hat Y  = X \hat \beta + \epsilon$$

- $E[\hat Y] = E[X\hat \beta] + 0 = X\beta$


- $Var[\hat Y] = Var[X \hat\beta + \epsilon] = X.Var[\hat \beta].X^T + Var[\epsilon]= X.\hat \sigma^2(X^TX)^{-1}.X^T + \hat \sigma^2 = H \hat \sigma^2 + \hat \sigma^2$


- $ \hat Y \sim N(X\beta, \space \space \hat \sigma^2.H + \hat \sigma^2)$


Prediction Interval Estimate:-
$$ \hat Y_j \space +- \space \space CriticalValue \space . \space SE(\hat Y_j)$$
$$ \hat Y_j \space +- \space \space t.stat_{n-p-1} . \sqrt{\hat \sigma^2H + \hat \sigma^2}$$


$$ Lower.bound: \hat Y_j - t.stat_{n-p-1} . \sqrt{\sigma^2.X_{train}.(X_{train}^T.X_{train})^{-1}.X_{train}^T + \sigma^2}$$
$$ Upper.bound: \hat Y_j + t.stat_{n-p-1} . \sqrt{\sigma^2.X_{train}.(X_{train}^T.X_{train})^{-1}.X_{train}^T + \sigma^2}$$

##### Confidence Interval Estimate

```r
#### BY HAND ####
# n-p-1
dof = nrow(train) - (length(coef(lm_marketing))-1) - 1

# level of significance
alpha = 0.05

# Training X Matrix
X = model.matrix(lm_marketing)

# Estimate of Variance
rss = sum(lm_marketing$residuals^2)
sigma2_hat = rss/dof

# Y_hat
y_hat = X %*% coef(lm_marketing)

# T-statistics
t_stat = qt(1-alpha/2, dof)

# C.I for 1st training prediction:
lwr.bound = y_hat[1] - t_stat * sqrt( sigma2_hat * X %*% solve(t(X) %*% X) %*% t(X) )[1,1]
upr.bound = y_hat[1] + t_stat * sqrt( sigma2_hat * X %*% solve(t(X) %*% X) %*% t(X) )[1,1]
cat("C.I for 1st: \tLower bound: ", lwr.bound, "\tC.I Upper Bound: ", upr.bound)

# C.I for 2nd training prediction:
lwr.bound = y_hat[2] - t_stat * sqrt( sigma2_hat * X %*% solve(t(X) %*% X) %*% t(X) )[2,2]
upr.bound = y_hat[2] + t_stat * sqrt( sigma2_hat * X %*% solve(t(X) %*% X) %*% t(X) )[2,2]
cat("C.I for 2nd: \tLower bound: ", lwr.bound, "\tC.I Upper Bound: ", upr.bound)

# C.I for 3rd training prediction:
lwr.bound = y_hat[3] - t_stat * sqrt( sigma2_hat * X %*% solve(t(X) %*% X) %*% t(X) )[3,3]
upr.bound = y_hat[3] + t_stat * sqrt( sigma2_hat * X %*% solve(t(X) %*% X) %*% t(X) )[3,3]
cat("C.I for 3rd: \tLower bound: ", lwr.bound, "\tC.I Upper Bound: ", upr.bound)
```

- $E[\hat Y^{\star}] = E[X^{\star}\hat \beta] + 0 = X^{\star}\beta$

Prediction Interval Estimate:-
$$ \hat Y^{\star}_j \space +- \space \space CriticalValue \space . \space SE(\hat Y^{\star}_j)$$
$$ \hat Y_j +- t.stat_{n-p-1} . \sqrt{\sigma^2(X(X^TX)^{-1}X^T + 1)}$$


$$ Lower.bound: \hat Y_j - t.stat_{n-p-1} . \sqrt{\sigma^2.X_{test}.(X_{train}^T.X_{train})^{-1}.X_{test}^T + \sigma^2}$$

```r
#### BY HAND ####
# n-p-1
dof = nrow(train) - (length(coef(lm_marketing))-1) - 1

# level of significance
alpha = 0.05

# Training X Matrix
X = model.matrix(lm_marketing)

# Testing X Matrix
X_test = cbind(1, data.matrix(test[,1:3]))

# Estimate of Variance
rss = sum(lm_marketing$residuals^2)
sigma2_hat = rss/dof

# Y_hat_star
y_hat_star = X_test %*% coef(lm_marketing)

# T-statistics
t_stat = qt(1-alpha/2, dof)

# C.I for 1st training prediction:
lwr.bound = y_hat_star[1] - t_stat * sqrt( sigma2_hat * X_test %*% solve(t(X) %*% X) %*% t(X_test) + sigma2_hat )[1,1]
upr.bound = y_hat_star[1] + t_stat * sqrt( sigma2_hat * X_test %*% solve(t(X) %*% X) %*% t(X_test) + sigma2_hat )[1,1]
cat("C.I for 1st: \tLower bound: ", lwr.bound, "\tC.I Upper Bound: ", upr.bound)

# C.I for 1st training prediction:
lwr.bound = y_hat_star[2] - t_stat * sqrt( sigma2_hat * X_test %*% solve(t(X) %*% X) %*% t(X_test) + sigma2_hat )[2,2]
upr.bound = y_hat_star[2] + t_stat * sqrt( sigma2_hat * X_test %*% solve(t(X) %*% X) %*% t(X_test) + sigma2_hat )[2,2]
cat("\nC.I for 2nd: \tLower bound: ", lwr.bound, "\tC.I Upper Bound: ", upr.bound)

# C.I for 1st training prediction:
lwr.bound = y_hat_star[3] - t_stat * sqrt( sigma2_hat * X_test %*% solve(t(X) %*% X) %*% t(X_test) + sigma2_hat )[3,3]
upr.bound = y_hat_star[3] + t_stat * sqrt( sigma2_hat * X_test %*% solve(t(X) %*% X) %*% t(X_test) + sigma2_hat )[3,3]
cat("\nC.I for 3rd: \tLower bound: ", lwr.bound, "\tC.I Upper Bound: ", upr.bound)
```

## 7.2. F-Test

At times, t-test is in-conclusive such as it may suggest that some of the predictors are significant. 

- For example, let's say we have 20 predictors, so there are high chances that it may suggest multiple predcitors as significant *i.e. p-value is extremely small*, but in actual scenario, since we are performing individual t-tests for all the predictors there are good chances some or the other will be a False Positive.


- Also, there could be Type I errors (False Positives) when we should actually reject the Null hypothesis, but we don't by mistake and accept it making some predictor as in-significant.


- Type I error e.g.: 

    - Chocolate = Beta * Weight_Loss
    - Null Hypothesis: Beta = 0
    - t-test: p-value > alpha
    - Fail to reject the Null Hyp, and Beta can be zero.



#### General Hypothesis:

Compares two models:

Let's define, one model by $A$ and other reduced model by $B$:

$A:$ $Y_i = \beta_0 + \beta_1X_{youtube} + \beta_2 X_{facebook} + \beta_3 X_{newspaper}$

$B$: $Y_i = \beta_0$

.

We state the hypothesis as:

$H_0:$ Reduced Model $B$ is sufficient to explain the variability in response

$H_a:$ Reduced Model $B$ is not sufficient and at-least one predictor is required

.

### $F-Stats$

$F(stat) = \frac{\frac {RSS_{reduced} - RSS_{full}}{p-q}}{\frac{RSS_{full}}{n-p-1}} \sim F.dist_{p-q,n-p-1}$

..

### 1). Initial Full F-Test:

Let’s define, the full model by A and reduced model containing just the intercept term, by B

$A:$ $Y_i = \beta_0 + \beta_1X_{youtube} + \beta_2 X_{facebook} + \beta_3 X_{newspaper}$

$B$: $Y_i = \beta_0$

We state the hypothesis as:

H0:
 Reduced Model B
 is sufficient to explain the variability in response

Ha:
 Reduced Model B
 is not sufficient and at-least one predictor is required

```r
full_model = lm(sales ~ youtube + facebook + newspaper, data=train)
full_summary = summary(full_model)

# F-Test Value
# by default summary() performs full F-test

full_summary$fstatistic
```

`F-statistic: 511.9 on 3 and 153 DF,  p-value: < 2.2e-16`

- As it is clear from the last line of the summary which performed the Full F-Test, the F-stat value is 511.9 and the p-value is very small, much less than the $\alpha$, thus we can reject the Null Hypothesis which stated that the reduced model is enough to explain the variability in the response Sales.


- Therefore, **the empty reduced model (model only with intercept) is not sufficient** and we need one/more predictors to explain the response variable's variability.

### 2). Partial F-Test 1:

Let's define, the full model by $A$ and partially-reduced model by $D$:

$A:$ $Y_i = \beta_0 + \beta_1X_{youtube} + \beta_2 X_{facebook} + \beta_3 X_{newspaper} + \varepsilon_i$

$D$: $Y_i = \beta_0 + \beta_1X_{youtube} + \beta_2 X_{newspaper} + \varepsilon_i$

**We state the hypothesis as:**

$H_0:$ Reduced Model $D$ is sufficient to explain the variability in response

$H_a:$ Reduced Model $D$ is not sufficient.

```r
reduced_model_1 = lm(sales ~ youtube + newspaper, data=train)

anova(reduced_model_1, full_model)
```

- Clear from ANOVA results, that the p-value of the reduced model with only two predictors is very small, suggesting that we reject the null hypothesis which stated that the reduced model (D) is sufficient to explain.

### 3). Partial F-Test 2:

Let's define, the full model by $A$ and partially-reduced model by $C$:

$A:$ $Y_i = \beta_0 + \beta_1X_{youtube} + \beta_2 X_{facebook} + \beta_3 X_{newspaper} + \varepsilon_i$

$C$: $Y_i = \beta_0 + \beta_1X_{youtube} + \beta_2 X_{facebook} + \varepsilon_i$

**We state the hypothesis as:**

$H_0:$ Reduced Model $C$ is sufficient to explain the variability in response

$H_a:$ Reduced Model $C$ is not sufficient.

```r
reduced_model_2 = lm(sales ~ youtube + facebook, data=train)

anova(reduced_model_2, full_model)
```

- Clear from ANOVA results, that the p-value of the reduced model with only two predictors is 0.4019 which is greater than level of significance of 0.05, suggesting that we fail to reject the null hypothesis and the reduced model (C) is sufficient to explain the variability in response, it is our final model!

Therefore, this is our final model.

### Shapiro Wilk Test


The Shapiro-Wilk’s test or Shapiro test is a normality test in frequentist statistics. The null hypothesis of Shapiro’s test is that the population is distributed normally. If the value of p is equal to or less than 0.05, then the hypothesis of normality will be rejected by the Shapiro test. On failing, the test can state that the data will not fit the distribution normally with 95% confidence. However, on passing, the test can state that there exists no significant departure from normality. This test can be done very easily in R programming.


$W=\frac{(\sum_{i=1}^n a_ix_{(i)})^2}{(\sum_{i=1}^n x_i - \bar{x})^2}$

where,

- x(i) : it is the ith smallest number in the given sample.
- mean(x) : ( x1+x2+……+xn) / n i.e the sample mean.
- ai : coefficient that can be calculated as (a1,a2,….,an) = (mT V-1)/C . 

Here V is the covariance matrix, m and C are the vector norms that can be calculated as C= || V-1 m || and m = (m1, m2,……, mn ).


###### Hypothesis:

$H_o:$ Samples residuals came from a normal distribution.

$H_a:$ Samples residuals did not came from a normal distribution.

```r
# Shapiro test of Normality

# loading the package
library("dplyr")

# Using the shapiro.test() to check
shapiro.test(df$youtube)
shapiro.test(df$facebook)
shapiro.test(df$newspaper)
```

- For all three features, the p-value is small and less than the alpha, thus we reject the Null hypothesis which stated these samples came from a normal distribution. 


- Thus, the hypothesis of normality is rejected by the Shapiro test.


.

***Normality assumption is violated!***

.

### Checking for Outliers, Leverage and Influential Points:


An outlier is a point with a large residual. Non-leverage outliers are non-influential points which are usual and differ from the common sense or notation of observations.

A leverage is again a unusual point but generally that is leverage or used to create regression fit.

An influential point is a point that has a large impact on the regression. A point can be an outlier without being influential.


Now, for outlier/leverage detection, there are three major ways:-

1. We can look at their leverage, which depends only on the value of the predictors.
2. We can look at their studentized residuals, either ordinary or cross-validated, which depend on how far they are from the regression line.

3. We can look at their Cook's statistics, which say how much removing each point shifts all the fitted values; it depends on the product of leverage and residuals.

## 3. Applying GLS

Durbin-Watson Statistics (Estimate of OMEGA $\Omega$ ):

We know that,
$$\varepsilon_{i+1} = {\rho}.\varepsilon_i + \gamma$$

Such that, 

- $\gamma \sim N(0, \tau^2)$


- Note that $\rho$ is the **correlation coefficient**


- We can estimate $\rho$ from the LR model (above). We caculated $\rho$ above as 0.583338985012419


Under the assumption:

$$\sigma = \rho^{|i - j|}$$
$=> \sigma = 0.58339^{|i - j|}$

```r
# diagonal matrix nxn (length of residuals 1 on diagonals and 0 elsewhere)
sigma = diag(n)    

corr_coeff = 0.58339

OMEGA = corr_coeff^abs(row(sigma) - col(sigma))
```

##### GLS : ######

Main Equation: $$Y^{\sim} = X^{\sim}\beta + V$$

where, $\sim$ means weighted versions:
- $Y^{\sim} = P^{-1}.Y$

- $X^{\sim} = P^{-1}.X$

- $V = P^{-1}.\epsilon$


And estimate of OMEGA,

$\Omega = \sigma^2 = P.P^T = P^2 = corr^{|i-j|} = \rho^{|i-j|}$

And from GLS derivation, we know:

$$\widehat \beta_{GLS} = (X^T.\Omega^{-1}.X)^{-1}X^T.\Omega^{-1}.Y \space\space\space\space\space\space\space\space\space\space\space\space\space\space\space;\space\space \widehat \beta_{GLS} \sim \space\space N \space (\space 0, \space\space(X^T.\Omega^{-1}.X)^{-1} \space\space\space)$$

GLS Estimate Equation: $$\widehat Y^{\sim} = X^{\sim}.\widehat \beta$$

### 3.1. GLS By hand

- Using mathematic formulation as above.

```r
##### GLS APPROACH 1 #####

# X: Design matrix
X = model.matrix(lmod)

# Y: Response
Y = df_globwarm$nhtemp

# Omega inverse
OMEGA_inv = solve(OMEGA)

# ESTIMATE OF BETA (GLS)
beta_hat_gls = solve( t(X) %*% OMEGA_inv %*% X ) %*% t(X) %*% OMEGA_inv %*% Y
beta_hat_gls

# Variance
beta_hat_gls_variance = solve( t(X) %*% OMEGA_inv %*% X )

# ^Y: Predicted Y
Y_hat = X %*% beta_hat_gls

# error (residuals) = Y - ^Y
residuals_gls = Y - Y_hat

# RSS
rss_gls = sum((Y - Y_hat)^2)
# ESS
ess_gls = sum((Y_hat - mean(Y))^2)
# TSS
tss_gls = sum((Y - mean(Y))^2)
# R2
rsquared = 1 - (rss/tss)
# MSE
mse = 1/n * sum((Y - Y_hat)^2)
# RMSE
rmse = sqrt(mse)
cat("\nMetrics:\n\nRSS:", rss, "\nESS:", ess, "\nTSS:", tss, "\nR2:", rsquared, "\nMSE:", mse, "\nRMSE:", rmse)
```

```r
# correlation between residuals lagged by one year, 

cor(residuals_gls[-1], residuals_gls[-n])
```

This correlation value for a lag of one year is not so different from the one we got earlier.

### 3.2. GLS By hand (Another way to apply GLS)

- Using mathematic formulation as above.

```r
##### GLS APPROACH 2 #####

# P is the square root of Sigma matrix (choleski decomp)
P <- chol(OMEGA)
P = t(P)

# P^-1
P_inv <- solve(P)

PX <- P_inv %*% X
Py <- P_inv %*% Y

# GLS:
# -1 because PX already has an intercept in it so -1 makes sure that no additional intercept is fitted
lmod_gls2 = lm(Py ~ PX -1)
lmod_gls2
```

We obtain the same values of coefficients!

# 10. Weighted Least Squares (WLS)

### 1. Dataset

```r
head(df, 5)
```

Suppose that the **Sales** repsonse is linked to the **predictor newspaper** in the following way:

$$Sales = a_0 . Newspaper^{a1}$$


- Perform simple linear regression on (transformations of) sales and newspaper to estimate a0 and a1.


- Use these estimates to perform weighted least squares where the weights are the inverse of the variance of Sales. 

### 2. Log transformation

$\log{Sales} = \log{a_0Newspaper^{a1}} = \log{a_0} + a_1.\log{Newspaper}$

```r
# Log-Transform the variables
# log(Sales) = log(a0) + a1.log(Newspaper)

sales.log = log(df$sales)
newspaper.log = log(df$newspaper)
```

### 3. Simple LR

```r
# Linear regression model using transformed variables

lmod_simple_log <- lm(sales.log ~ newspaper.log)
summary(lmod_simple_log)
```

- Residual standard error: 0.4101 on 198 degrees of freedom

### 4. Estimated coefficients $a_0, a_1$

$a_0 = logInverse(est.a_0)$

$a_1 = est.a_1$

```r
# intercept estimate   
a0 = exp(lmod_simple_log$coefficients[1])          # a0 = log-inverse(estimate); log-inverse in R is exp()

# slope estimate
a1 = lmod_simple_log$coefficients[2]

cat("Estimates: \ta0 =", a0, "\ta1 =", a1)
```

### 5. WLS

```r
# Weights for the weighted least squares regression
weights = 1 / (a0 * newspaper.log^a1)

# perform weighted least squares regression
wls_model = lm(sales.log ~ newspaper.log, data=df, weights=weights)

# view summary of model
summary(wls_model)
```

- The new Residual standard error is 0.1117 which is an improvement over the previous SLR model.


This indicates that the predicted values produced by the weighted least squares model are much closer to the actual observations compared to the predicted values produced by the simple linear regression model.

# 10. Model Selection

Two types of approaches:

1. Testing Methods: Backward elimination, forward elimination, Step-wise Regression
    - 1.1. Backward Elimination: Start with the full model, drop features if their p-value is above a threshold set.
    - 1.2. Forward Elimnation: Start with a simple LR model (choose one feature with lowest p-value) and add features untill all features have a p-value below a threshold set.
    - 1.3. Stepwise Regression: Add/drop features on the go.



2. Criterion Methods: AIC, BIC, Adjusted R2, MSPE
    - 2.1. $AIC = 2(p+1) - log L(Beta) = 2(p+1) + n log(RSS/n)$;   Lower the value, better the model.
    - 2.2. $BIC = log(n)(p+1) - log L(Beta) = log(n)(p+1) + n log(RSS/n)$;   Lower the value, better the model
    - 2.3. $Adj R2 = 1 - RSS/(n-p-1) / TSS/(n-1) = 1 - RSS * (n-1) / TSS * (n-p-1)$;   Higher the value, better!
    - 2.4. MSPE = 1/n SUM(y - y^)^2;     Lower the value better the model!

```r
# library(RCurl) #a package that includes the function getURL(), which allows for reading data from github.
# url = getURL(paste0("https://raw.githubusercontent.com/bzaharatos/",
#                     "-Statistical-Modeling-for-Data-Science-Applications/",
#                     "master/Modern%20Regression%20Analysis%20/",
#                     "Datasets/fabricsoundabsorb.csv"))

df_fabric = read.csv("Data/fabricsoundabsorb.csv", sep = ",")
head(df_fabric)
```

```r
## LR

y = df_fabric$acoustic1

lm_fabric = lm(acoustic1 ~ thickness + diameter + perforation + weight + stiffness + airPerm, data=df_fabric)
summary(lm_fabric)
```

- Multiple features have a higher p-value (thus, FTR Null Hypothesis, and feature parameter is insignificant, 0)


- We can remove some, but should we remove all of these ??

### 10.1. Backward Elimination

- Set $\alpha_0$ called as `p-to-remove` as a threshold.


- Repeat the process of elimination untill all features are below this threshold.


- Remove the feature with highest p-value!

```r
# Setting p-to-remove, alpha0 = 0.15
# p-values above this needs to be dropped!
```

```r
lm_fabric_backward = update(lm_fabric, .~. - thickness)
summary(lm_fabric_backward)
```

```r
lm_fabric_backward = update(lm_fabric_backward, .~. - diameter)
summary(lm_fabric_backward)
```

```r
lm_fabric_backward = update(lm_fabric_backward, .~. - stiffness)
summary(lm_fabric_backward)
```

- This is our final model

### 10.2. Forward Elimination

- Set $\alpha_0$ called as `p-to-remove` as a threshold.


- Repeat the process of addition untill some features are added, and are below this threshold.


- Choose the features based on highest p-value!

```r
# Setting p-to-remove, alpha0 = 0.15
# p-values above this needs to be dropped!

summary(lm_fabric)
```

```r
# Choose the feature with lowest p-value to start with

lm_fabric_forward = lm(acoustic1 ~ airPerm, data=df_fabric)
summary(lm_fabric_forward)
```

```r
# Choose the feature with lower p-value to continue

lm_fabric_forward = update(lm_fabric_forward, . ~ . + weight)
summary(lm_fabric_forward)
```

```r
# Choose the feature with lower p-value to continue

lm_fabric_forward = update(lm_fabric_forward, . ~ . + perforation)
summary(lm_fabric_forward)
```

- There are no other features having a p-value lower than our alpha0, thus, this is the final model!

### 10.3. AIC

The `regsubsets()` function in the `leaps` library can fit models with different number of predictors and select the best model!

$AIC = 2(p+1) + n\log{\frac{RSS}{n}}$

```r
library(leaps)
```

```r
fit_all_models_at_once = regsubsets(acoustic1 ~ thickness + diameter + perforation + weight + stiffness + airPerm, data = df_fabric)
result = summary(fit_all_models_at_once)

# print which features were fitted during fitting all of the models
result$which    # or use, result$outmat
```

- Tells that the first model was fitted using intercept and airPerm.
- #2: lm(acoustic1 ~ airPerm + thickness), likewise.

```r
n = nrow(df_fabric)

# p     = 1, 2, 3, 4, 5, 6
# p+1   = 2, 3, 4, 5, 6, 7
p_plus_1 = c(2, 3, 4, 5, 6, 7)

AIC = 2*(p_plus_1) + n*log(result$rss / n)
plot(AIC, xlab='Number of predictors', ylab="AIC")
```

- Best model is with 3 predictors.
- Best is: `lm(acoustic1 ~ airPerm + weight + perforation)`

### 10.4. BIC

The `regsubsets()` function in the `leaps` library can fit models with different number of predictors and select the best model!

```r
BIC = result$bic
plot(BIC, xlab='Number of predictors', ylab="BIC")
```

- Best model is with 3 predictors.
- Best is: `lm(acoustic1 ~ airPerm + weight + perforation)`

# 11. Problem of Multicollinarity

- A psuedo 5th assumption of linear regression is the collinarity between predictors.


- For the GRAM matrix inverse to exist, such that OLS produces unique solution, we need to make sure there is no or very low correlation between predictors.


    - No/low correlation between predictors.
    - One feature should not be a **NEAR** linear combination of other predictors.
    
    
    
- Diagnosis:
    - 1. Correlation Matrix for pair-wise correlation between predictors.
    - 2. Value of estimated parameters and sign of SE of estimate of parameters.
    - 3. Variance Inflation Factor (VIF)
        - Value Ranges:
            - VIF < 5: No Multicollinarity!
            - 5 < VIF < 10: More Checks needed to prove Multicollinarity!
            - VIF > 10: Problem of Multicollinarity!
    - 4. Condition Number (Kappa)

```r
# Same Fabric Dataset

head(df_fabric, 3)
```

```r
lm_fabric_full = lm(acoustic1 ~ thickness + diameter + perforation + weight + stiffness + airPerm, data=df_fabric)
lm_fabric_reduced = lm(acoustic1 ~ perforation + weight + airPerm, data=df_fabric)
```

### 1. Multicollinarity Diagnosis: Correlation Matrix

- -1: Strong negative correlation
- 0: No correlation
- +1: Strong positive correlaiton

```r
corrplot(cor(df_fabric), method = "ellipse", addCoef.col = "black", tl.col = "black")
```

- There are some pairwise collinarity that could contribute to Multicollinarity problem!

```r
# let's see for the best selected model based on AIC
```

```r
corrplot(cor(df_fabric[, c('airPerm', 'weight', 'perforation')]), method = "ellipse", addCoef.col = "black", tl.col = "black")
```

- As we can see, there is multicollinarity problem between `airPerm` and `perforation` features, even in the best model

### 2. Multicollinarity Diagnosis: Value of estimated parameters and sign of SE of estimate of parameters

```r
summary(lm_fabric_full)
```

- Need domain understanding, that whether the estimate of airPerm should be so low and negative?

### 3. Multicollinarity Diagnosis: VIF

- 0 < VIF < 5: No Multicollinarity!
- 5 < VIF < 10: More Checks needed to prove Multicollinarity!
- VIF > 10: Problem of Multicollinarity!

```r
# full model VIF:
data.frame(vif(lm_fabric_full))
```

- As we can see, features like weight and thickness should be dropped since they have high VIF values!

```r
# full model VIF:
data.frame(vif(lm_fabric_reduced))    # or any reduced model check
```

- This looks great! All VIF values are less than 5.

### 4. Multicollinarity Diagnosis: Kappa

```r
kappa(lm_fabric_full)
```

```r
kappa(lm_fabric_reduced)
```

- Condition number is still too high, for both versions, hence more research should be carried out!
